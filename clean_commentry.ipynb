```python
deliveries=['Bouncer','Inswinger','Reverse swing','Leg cutter','Off cutter','Outswinger','Yorker','Beamer','Knuckleball','Slower ball','full','short']
def ball_type(comment):
    for ball in deliveries:
        if ball.casefold() in comment:
            print(ball)
a=df['Commentary Text'][40]
ball_type(a)   
```

```python
balls=[]
def runs(comment):
    extra=None
    run=comment.split(',')[1]
    balls.append(run)
    print(run)
    if '1 run'.casefold() in run:
        run = 1
    elif '2 runs'.casefold() in run:
        run= 2
    elif 'no run'.casefold() in run:
        run= 0
    elif '3 runs'.casefold() in run:
        run = 3
    elif 'Four'.casefold() in run:
        run =4
    elif 'Six'.casefold() in run:
        run =6
    elif 'byes'.casefold() in run:
        run=comment.split(',')[2]
    elif 'out'.casefold() in run:
        run='Out'
    elif 'wides'.casefold() in run:
        if run.split(' ')[1]:
            extra=run.split(' ')[1]
            run='Wide'
        else:
            run='Wide'
    
    
    return run,extra
for j,i in enumerate(commentry):
    print(j,list(runs(i)))
```
